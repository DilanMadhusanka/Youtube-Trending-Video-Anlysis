```python
items = res['items']
```

```python
for i in range(0,10):
    kind.append((items[i])['kind'])
    etag.append((items[i])['etag'])
    videoId.append((items[i])['id'])
    publishedAt.append((items[i])['snippet']['publishedAt'])
    channelId.append((items[i])['snippet']['channelId'])
    title.append((items[i])['snippet']['title'])
    description.append((items[i])['snippet']['description'])
    
    reqStatic = youtube.videos().list(id=(items[i])['id'], part='statistics', regionCode='LK')
    reqStatic = reqStatic.execute()
    
    viewCount.append(reqStatic['items'][0]['statistics']['viewCount'])
    likeCount.append(reqStatic['items'][0]['statistics']['likeCount'])
    favoriteCount.append(reqStatic['items'][0]['statistics']['favoriteCount'])
    commentCount.append(reqStatic['items'][0]['statistics']['commentCount'])
#     dislikeCount.append(reqStatic['items'][0]['statistics']['dislikeCount'])
    
    reqComments = youtube.commentThreads().list(part='snippet', videoId=(items[i])['id'])
    resComments = reqComments.execute()
#     print(resComments['items'][0]['snippet']['topLevelComment']['snippet']['textDisplay'])
#     print(resComments['items'][0]['snippet']['topLevelComment']['snippet']['textOriginal']+"---------")
    comment.append(resComments['items'][0]['snippet']['topLevelComment']['snippet']['textOriginal'])

    reqChannel = youtube.channels().list(part='snippet', id=(items[i]['snippet']['channelId']))
    resChannel = reqChannel.execute()
#     print(resChannel['items'][0]['snippet']['title'])
    channelTitle.append(resChannel['items'][0]['snippet']['title'])
    
    reqVideoCategory = youtube.videoCategories().list(part='snippet', id=(items[i])['snippet']['categoryId'])
    resVideoCategory = reqVideoCategory.execute()
    videoCategory.append(resVideoCategory['items'][0]['snippet']['title'])
    
#     print(resVideoCategory['items'][0]['snippet']['title'])
    
    i+=1
    
```

```python
data={
    'kind':kind,
    'etag':etag,
    'trending_date':pd.to_datetime("today").date(),
    'videoId':videoId,
    'publishedAt':publishedAt,
    'channelId':channelId,
    'channelTitle':channelTitle,
    'title':title,
    'description':description,
    'comment':comment,
    'viewCount':viewCount,
    'likeCount':likeCount,
    'videoCategory':videoCategory,
    'favoriteCount':favoriteCount,
    'commentCount':commentCount
}

df=pd.DataFrame(data)
df.head()
```

```python
df.to_csv('treding_video.csv', index=False)
```
